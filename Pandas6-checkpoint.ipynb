```python
from wordcloud import WordCloud,STOPWORDS
import matplotlib.pyplot as plt 
```

```python
stopwords=set(STOPWORDS)
print(stopwords)
```

```python
wc=WordCloud(width=800,height=400,max_words=50,background_color='white')
s='Straw Hat captain is Luffy. Luffy has master three haki in One Piece. One Piece is a hidden treasure. Luffy right wing man is Zoro. Luffy left wing man is Sanji'
wc.generate(s)
plt.imshow(wc)
plt.axis('off')
plt.show()
```

```python
s=open('speech.txt').read()
wc.generate(s)
plt.imshow(wc)
plt.axis('off')
plt.show()
```

```python
import re 
```

```python
s='The rain in spain 123'
x=re.findall(r'[a-m ,][n-z]',s)
print(x)
print(len(x))
```

```python
x=re.findall(r'[a-m1-9]',s)
print(x)
print(len(x))
```

```python
s="The rain in spain 123"
x=re.findall(r'[^a-m]',s)
print(x)
print(len(x))
```

```python
s="The rain in spain 123"
x=re.findall(r'^T.*3$',s)
print(x)
print(len(x))
```

```python
s='hello helo helo heo ho h o'
x=re.findall(r'h.{1,4}o',s)
print(x)
print(len(x))
```

```python
s='hello helo helo heo ho h o'
x=re.findall(r'h|o',s)
print(x)
print(len(x))
```

```python
s='black, blue green Black'
x=re.findall(r'\bbl\w{3}',s)
print(x)
print(len(x))
```

```python
s='black, blue green Black 3456'
x=re.findall(r'\D',s)
print(x)
print(len(x))
```

```python
s='black, blue green Black 3456'
x=re.findall(r'bl',s)
print(x)
print(len(x))
```

```python
s='black, blue green Black 3456'
x=re.search(r'bl',s)
print(x.group())
# print(len(x))
```

```python
s='black, blue green Black 3456'
x=re.split(r'bl',s,1)
print(x)
print(len(x))
```

```python
s='black, blue green Black 3456'
x=re.sub(r'bl','Vishal10',s,1)
print(x)
print(len(x))
```
