```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
d={'x':[1,2,3,4,5,6],'y1':[10,30,40,50,70,80],'y2':[120,90,100,150,20,10]}
df=pd.DataFrame(d)
print(df)
plt.stackplot(df.x,df.y1,df.y2,labels=["y1","y2"])
plt.title("Nilesh")
plt.legend()
plt.show()
```

```python
plt.fill_between(df.x,df.y1,df.y2,color='yellow')
plt.plot(df.x,df.y1,'o:b')
plt.plot(df.x,df.y2,'v--r')
plt.show()
```

```python
plt.fill_between(df.x,df.y1,df.y2,where=df.y2>df.y1,color='yellow',interpolate=True)#here where is used for color pupose
plt.fill_between(df.x,df.y1,df.y2,where=df.y2<df.y1,color='green',interpolate=True)
plt.show()
```

```python
d={'x':[10,30,60,70,80,120],'y1':[10,30,40,50,70,80],'y2':[120,90,100,150,20,10]}
df=pd.DataFrame(d)
print(df)
plt.scatter(df.x,df.y1,color=['red','yellow','green','blue','pink','purple'])
plt.scatter(df.x,df.y2)
plt.show()
```

```python
fig,ax=plt.subplots()
ax.scatter(df.x,df.y1,color=['red','yellow','green','blue','pink','purple'],marker='v',s=30)
ax.scatter(df.x,df.y2,marker='x',color='black',s=70)
ax.set_xticks(df.x)
plt.show()
```

```python
df=pd.read_csv("auto-mpg.csv")
print(df)
```

```python
plt.scatter(df.mpg,df.acceleration)
plt.show()
```

```python
plt.boxplot(df.mpg,notch=True,widths=0.7)
plt.show()
```

```python
import seaborn as sns
```

```python
df.boxplot('mpg',notch=True,widths=0.75,by='cylinders')
plt.show()
```

```python
sns.boxplot(x="cylinders",y="mpg",data=df)
plt.show()
```

```python
sns.regplot(x=df.mpg,y=df.acceleration,marker="x")
plt.show()
```

```python
sns.regplot(x=df.mpg,y=df.acceleration,marker="x",scatter_kws={"color":"red","alpha":0.7})
plt.show()
```

```python
sns.regplot(x=df.mpg,y=df.acceleration,marker="x",scatter_kws={"color":"red","alpha":0.7},
            line_kws={'color':'green','linewidth':10,'ls':':'})
plt.show()
```

```python
data=np.random.randint(low=1,high=100,size=(10,10))
print(data)
```

```python
sns.heatmap(data=data,annot=True)
plt.show()
```

```python
sns.heatmap(data=data,annot=True,vmax=60,vmin=40)
plt.show()
```

```python
sns.heatmap(data=data,annot=True,vmax=60,vmin=40,cmap='tab20')
plt.show()
```

```python
sns.heatmap(data=data,annot=True,vmax=60,vmin=40,cmap='tab20',linewidth=2,linecolor='blue')
plt.show()
```

```python
sns.heatmap(data=data,annot=True,vmax=60,vmin=40,cmap='tab20',linewidth=2,linecolor='blue',cbar=False)
plt.show()
```

```python
sns.heatmap(data=data,annot=True,vmax=60,vmin=40,cmap='coolwarm',linewidth=2,linecolor='blue')
plt.show()
```

```python
from pywaffle import Waffle
```

```python
d={'c':['A','B','C','D'],'v':[20,30,40,50]}
df=pd.DataFrame(d)
fig=plt.figure(FigureClass=Waffle,rows=10,values=df['v'],labels=list(df['c']),legend={'loc':'upper left'})
plt.title('B6 Is Best')
plt.show()
```

```python
d={'c':['A','B','C','D'],'v':[20,30,40,50]}
df=pd.DataFrame(d)
fig=plt.figure(FigureClass=Waffle,rows=10,values=df['v'],labels=list(df['c']),legend={'loc':'upper left'},characters='😀')
plt.title('B6 Is Best')
plt.show()
```
