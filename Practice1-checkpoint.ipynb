```python
Q1=df.Fare.quantile(0.25)
Q3=df.Fare.quantile(0.75)
IQR=Q3-Q1
UL=Q3+1.5*IQR
LL=Q1-1.5*IQR
df=df[(df.Fare>=LL)&(df.Fare<=UL)]
print(df)
```

```python
L=[]
for i in df.Age:
    if i<18:
        L.append("children")
    elif i<30:
        L.append("young")
    elif i<50:
        L.append("young adult")
    elif i<100:
        L.append("senior")
print(L)
df.AgeGroup=np.array(L)
print(df.AgeGroup)
pd.crosstab(df.AgeGroup,df.Survived)
```
