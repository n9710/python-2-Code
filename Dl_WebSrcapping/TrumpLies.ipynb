```python
import pandas as pd 
import numpy as np
import requests 
from bs4 import BeautifulSoup
```

```python
with open("TrumpLies.html", "r", encoding = "utf-8") as f:
    data = f.read()
```

```python
soup = BeautifulSoup(data)
soup
```

```python
print("Q1 : Title : ")
title = soup.find("title").text.strip()
title
```

```python
print("Q2 : Footer Date : ")
# footer = soup.find("footer", class_ = "m-statement__footer").text.strip().split("😊")
footer = soup.find("footer").text.strip().split("😊")[1]
footer
```

```python
print("Q3 : Secreate Word")
text = soup.find("p", class_ = "secret-word").text
text
```

```python
print("Q4 : H1 And H2 Tag")
h1 = soup.findAll("h1")[0].text.strip()
h1_2 = soup.findAll("h1")[1].text.strip().replace("\n", "").replace(" ", "")
h2 = soup.findAll("h2")
print(h1,h1_2)
for i in h2:
    print(i.text.strip())
```

```python
print("Q5 : All Links : ")
links = soup.findAll("a")
href = []
for i in links:
    if i["href"].startswith("http"):
         href.append(i["href"])
print(len(href))
href
```

```python
print("Q6 : Records")
records = soup.find_all("span", class_ = "short-desc") #findAll same as find_all
print(len(records))
```

```python
print("Q7 : Records as given : ")
data = { "Date" : [], "Truth" : [], "Statement" : []}
for i in soup.findAll("span", class_ = "short-desc"):
    date = i.text.split("\n")[2].strip()
    data["Date"].append(date)
    truth = i.text.strip().split("“")[1].split("”")[0]
    data["Truth"].append(truth)
    statement = i.text.split("(")[1].split(")")[0]
    data["Statement"].append(statement)
    print(" Date : ",date )
    print(" Truth : ",truth)
    print(" Statement : ",statement)


```
