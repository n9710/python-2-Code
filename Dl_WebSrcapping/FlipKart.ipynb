```python
import pandas as pd
import numpy as np 
from bs4 import BeautifulSoup
```

```python
with open("newflipkart.html", "r", encoding = "utf-8") as f:
    data = f.read()
soup = BeautifulSoup(data)
```

```python
data = {"Name" : [], "Memory" : [], "Color" : [], "Stars" : [], "Reviews" : [], "Old Price" : [], "New Price" : [], "Discount" : []}
```

```python
for i in soup.findAll("div", class_ = "_75nlfW"):
    name = i.find("div", class_ = "KzDlHZ").text.split("(")[0].strip()
    memory = i.find("div", class_ = "KzDlHZ").text.split(", ")[1].split(")")[0]
    color = i.find("div", class_ = "KzDlHZ").text.split("(")[1].split(",")[0]
    star = i.find("div", class_ = "XQDdHH").text
    review = i.find("span", class_ = "Wphh3N").text.split("\xa0")[-1].split(" ")[-2]
    old = int(i.find("div", class_ = "yRaY8j ZYYwLA").text.split("₹")[-1].replace(",",""))
    new = int(i.find("div", class_ = "Nx9bqj _4b5DiR").text.split("₹")[-1].replace(",",""))
    dis = i.find("div", class_ = "UkUFwK").text.split(" ")[0]
    data["Name"].append(name)
    data["Memory"].append(memory)
    data["Color"].append(color)
    data["Stars"].append(star)
    data["Reviews"].append(review)
    data["Old Price"].append(old)
    data["New Price"].append(new)
    data["Discount"].append(dis)
df = pd.DataFrame(data)
df.drop_duplicates(inplace=True)
```
